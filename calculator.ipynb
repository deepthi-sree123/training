```python
x = int(input())
y = int(input())

def add(x,y):
    return x+y

def sub(x,y):
    return x-y

def mul(x,y):
    return x*y

def div(x,y):
    if y == 0:
        return "Error"
    return x/y

user = input("Select an operation (Add/Sub/Mul/Div) or Exit: ").title()

result = None

if user == "Add":
    result = add(x,y)
elif user == "Sub":
    result = sub(x,y)
elif user == "Mul":
    result = mul(x,y)
elif user == "Div":
    result = div(x,y)
elif user == "Exit" :
    result = "Program stopped"    
else:
  result = "Error"
  


print(result)
```
